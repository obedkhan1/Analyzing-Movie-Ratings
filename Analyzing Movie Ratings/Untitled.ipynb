# Data Cleaning:

```python
import pandas as pd

data = pd.read_csv('movie_data.txt')
pd.DataFrame(data)
data
```

```python
handleNull = data.dropna(subset=['Rating', 'Age'])
```

```python
handleNull
```

```python
handleNull.dtypes
```

```python
handleNull.loc[:, 'Age'] = handleNull['Age'].astype(int)

handleNull
```

```python
handleNull.dtypes
data1 = handleNull
data1
```

# Exploratory Data Analysis:

```python
data1.loc[:, 'Rating'] = pd.to_numeric(data1['Rating'], errors='coerce')

data1.loc[:, 'average_rating'] = data1.groupby('MovieName')['Rating'].transform('mean')
data1
```

```python
data1.loc[:, 'RatingCount'] = data1.groupby('MovieName')['Rating'].transform('size')
data1
```

```python
# Find the index of the most rated movie
most_rated_index = data1['RatingCount'].idxmax()

# Find the index of the least rated movie
least_rated_index = data1['RatingCount'].idxmin()

# Assign the indices to new columns
data1.loc[:, 'most_rated_movie'] = most_rated_index
data1.loc[:, 'least_rated_movie'] = least_rated_index

data1
```

```python
data1.drop(columns=['high'], inplace=True)
data1
```

# Visualization

```python
import matplotlib.pyplot as plt

# Plot the distribution of ratings
plt.hist(data1['Rating'], bins=5, edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python

top_n_movies = ['The Godfather', 'Inception', 'The Matrix', 'Parasite', 'Spirited Away']
ratings = [5, 4, 5, 4, 4] 

plt.barh(top_n_movies, ratings, color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Movie')
plt.title('Top-N Recommended Movies')
plt.show()
```
